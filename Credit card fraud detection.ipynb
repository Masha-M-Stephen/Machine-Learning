Modeling using Sklearn

```python
#Variables 
X = creditcard.drop(['Class'], axis =1)
Y = creditcard['Class']
print(X.shape)
print(Y.shape)
```

```python
#Standardize data
from sklearn import preprocessing
X =preprocessing.StandardScaler().fit(X).transform(X.astype(float))
X[0:5]
```

```python
#Creating a train and test set
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size =0.25, random_state = 2)

print('train set: ', X_train.shape, Y_train.shape)
print('test set: ', X_test.shape, Y_test.shape)
```

Algorithims used in modeling.

Model 1: Logistic regression 

Model 2: Random forest

Model 3: support vector machine


```python
#Model 1: Logistic regression --> estimating probabilities
from sklearn.linear_model import LogisticRegression
#model
logmodel = LogisticRegression()
#fit training
logmodel.fit(X_train, Y_train)
```

```python
#Logistic regression testing predictions
Ypred = logmodel.predict(X_test)
Ypred
```

```python
#Evalutation of Logistic model
from sklearn.metrics import f1_score, classification_report
#The score 
score = f1_score(Y_test, Ypred, average = "weighted")
score
```

```python
#Classification report
print(classification_report(Y_test, Ypred))
```

```python
#Confusion matrix
from sklearn.metrics import confusion_matrix, plot_confusion_matrix
cnf = confusion_matrix(Y_test, Ypred, labels = [1,0])
np.set_printoptions(precision=2)

cnf
```

Logistic model predicted:

77 true Fraud cases

71086 true Legit cases

30 casses predicted as Fraud but are Legit

9 cases predicted as Legit but are Fraud


```python
#Model 2: Random forest classifier
from sklearn.ensemble import RandomForestClassifier
#Using a gaussian classifier Model
rndmodel = RandomForestClassifier(n_estimators = 100)
#Fit training
rndmodel.fit(X_train, Y_train)
```

```python
#Random forest test predictions
Yhat =rndmodel.predict(X_test)
Yhat
```

```python
from sklearn import metrics
#Evaluating random forest model
#Evaluating metrics have been imported from the logistic regression
#Model accuracy
print('model accuracy: ', metrics.accuracy_score(Y_test,Yhat))
```

```python
#Classification report
print(classification_report(Y_test, Yhat))
```

```python
#confusion matrix
confusion_matrix, plot_confusion_matrix
cnf = confusion_matrix(Y_test, Yhat, labels = [1,0])
np.set_printoptions(precision=2)

cnf
```

Random forest model predictions:

88 true Fraud cases

71088 true Legit cases

19 casses predicted as Fraud but are Legit

7 cases predicted as Legit but are Fraud

The random forest model performes better than logistic model in this instance.

```python
#Model 3: support vector machine
from sklearn import svm
#Model
svmmodel = svm.SVC(kernel = 'linear')
#Fit training
svmmodel.fit(X_train, Y_train)
```

```python
#Model test predictions
Y_pred = svmmodel.predict(X_test)
```

```python
#Evaluating the model
print('model accuracy: ', metrics.accuracy_score(Y_test, Y_pred))
```

```python
#Classification report
print(classification_report(Y_test, Y_pred))
```

```python
#confusion matrix
cnf = confusion_matrix(Y_test, Y_pred, labels = [1,0])
np.set_printoptions(precision=2)
cnf
```
